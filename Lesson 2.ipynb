```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv("https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv")
```

```python
students_performance.head()
```

Выберем все строки `writing score > mean`

```python
mean_writing_score = students_performance["writing score"].mean()
```

```python
students_performance.loc[students_performance["writing score"] > mean_writing_score]
```

Выберем все строки, где `gender == female`

```python
students_performance.loc[students_performance.gender == "female"]
```

### Создадим результирующее условие

```python
command = (students_performance.gender == "female") & (students_performance["writing score"] > mean_writing_score)
```

```python
students_performance.loc[command].head()
```
