```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv("StudentsPerformance.csv")
```

```python
students_performance.head()
```

###  Фильтруем названия столбцов

```python
students_performance = students_performance.rename(columns={
    "parental level of education": "parental_level_of_education",
    "test preparation course": "test_preparation_course",
    "math score": "math_score",
    "reading score": "reading_score",
    "writing score": "writing_score"
})
```

```python
students_performance.head()
```

### Query

```python
students_performance.query("writing_score > 74").head()
```

```python
students_performance.query("gender == 'female'").head()
```

```python
students_performance.query("gender == 'female' & writing_score > 78").head()
```

### Query и переменные

```python
writing_score_query = 90
students_performance.query("writing_score > @writing_score_query")
```

### Работа с колонками
Не пандасовский метод:

```python
score_cols = [i for i in list(students_performance) if "score" in i]
students_performance[score_cols].head()
```

Пандасовский метод:

```python
students_performance.filter(like="score").head()
```
